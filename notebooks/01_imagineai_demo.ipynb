## Understanding the Core Math

Before we run imagineAI, let's understand what it's doing:

### The Φ Field
Every word/sentence becomes a point in high-dimensional space (embeddings).

### The Operators
- **∇Φ (Gradient)**: Direction from question toward answer
- **∇×F (Curl)**: Context memory / conversation loops  
- **∇²Φ (Laplacian)**: Whether meaning is expanding or focusing
- **σ (Residue)**: Semantic distance that must be minimized
- **ρ_q (Boundary)**: Constraints the answer must satisfy

```python
# Quick demo of the Φ field
import numpy as np
from imagine_ai.language.embeddings import get_embedding_backend
from imagine_ai.core.phi_field import PhiField

# Load embeddings
print("Loading embedding model...")
backend = get_embedding_backend("sentence-transformers")
phi = PhiField(backend)

# Embed some concepts
concepts = [
    "What is the capital of Mississippi?",
    "Jackson is a city in Mississippi",
    "The weather is nice today",
    "Python is a programming language"
]

states = [phi.embed(c) for c in concepts]

# Show distances
print("\nSemantic distances (lower = more similar):")
for i, s1 in enumerate(states):
    for j, s2 in enumerate(states):
        if i < j:
            dist = s1.distance_to(s2)
            print(f"  '{concepts[i][:40]}...' <-> '{concepts[j][:40]}...': {dist:.3f}")
```

## The Resolution Process

When you ask a question:
1. Question excites the Φ field
2. Boundary conditions (ρ_q) are extracted from the question
3. Knowledge is retrieved (Wikipedia)
4. LLM generates a candidate response
5. ITT operators evaluate: Does this minimize σ? Does it satisfy ρ_q?
6. If yes → answer. If no → iterate.

```python
# Demo the operators
from imagine_ai.core.operators import SemanticOperators, compute_sigma

ops = SemanticOperators(phi)

# Question and potential answer
question = phi.embed("What is the capital of Mississippi?")
correct_answer = phi.embed("Jackson is the capital of Mississippi")
wrong_answer = phi.embed("The weather is nice today")

# Compute gradients
grad_correct = ops.gradient(question, correct_answer)
grad_wrong = ops.gradient(question, wrong_answer)

print("Gradient Analysis:")
print(f"  Question → Correct answer: magnitude = {grad_correct.magnitude:.3f}")
print(f"  Question → Wrong answer: magnitude = {grad_wrong.magnitude:.3f}")

# Compute σ (residue)
sigma_correct = compute_sigma(question, correct_answer)
sigma_wrong = compute_sigma(question, wrong_answer)

print(f"\nResidues (σ):")
print(f"  Correct answer σ = {sigma_correct:.3f}")
print(f"  Wrong answer σ = {sigma_wrong:.3f}")
print(f"\nThe correct answer has LOWER σ - it resolves the field tension.")
```

## Full imagineAI Chat

Now let's use the full system:

```python
from imagine_ai.api.chat import ImagineAI

# Initialize (this may take a moment to load models)
print("Initializing imagineAI...")
ai = ImagineAI(
    embedding_backend="sentence-transformers",
    generator_type="simple",  # Use simple for testing without GPU
    use_wikipedia=True,
    verbose=True
)
print("Ready!")
```

```python
# Ask a question
question = "What is the capital of Mississippi?"
print(f"Question: {question}")
print()

answer = ai.chat(question)
print(f"\nAnswer: {answer}")
```

```python
# Try more questions
questions = [
    "What is the largest planet?",
    "What is the speed of light?",
    "Tell me about Jupiter"
]

for q in questions:
    print(f"Q: {q}")
    a = ai.chat(q)
    print(f"A: {a}")
    print()
```

```python
# Check conversation statistics
print("Session stats:")
stats = ai.get_stats()
for key, value in stats.items():
    print(f"  {key}: {value}")
```

## Understanding What Just Happened

The key difference from a regular LLM:

1. **Standard LLM**: `Input → Neural Network → Sample from probability → Output`
   - The output is whatever has high probability
   - No explicit constraints
   - Can hallucinate

2. **imagineAI**: `Question → Φ Excitation → σ-minimization under ρ_q → Stable Answer`
   - The output is the ONLY configuration where σ → 0
   - Explicit boundary conditions
   - Answer must satisfy constraints to be valid

The LLM is just the "field generator" - it proposes candidates. The ITT operators decide if those candidates actually resolve the question.
